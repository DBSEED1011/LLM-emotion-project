```python
import json
import os
import re

# 定义文件夹路径和文件名
folder_path = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/prompt"
# 定义文件名模式
pattern = re.compile(r"(\d+)_character\.json$")

# 获取文件夹中的所有文件
files = os.listdir(folder_path)

# 创建一个字典来存储文件对
file_pairs = {}

# 遍历文件夹中的文件，找到匹配的文件对
for file_name in files:
    match = pattern.match(file_name)
    if match:
        prefix = match.group(1)
        character_file = file_name
        setting_file = f"{prefix}_game_setting_prompt.json"
        if setting_file in files:
            file_pairs[prefix] = (character_file, setting_file)

# 检查每个文件对中的id是否一致
for prefix, (file1_name, file2_name) in file_pairs.items():
    file1_path = os.path.join(folder_path, file1_name)
    file2_path = os.path.join(folder_path, file2_name)

    # 读取两个JSON文件
    with open(file1_path, "r", encoding="utf-8") as file1:
        data1 = json.load(file1)

    with open(file2_path, "r", encoding="utf-8") as file2:
        data2 = json.load(file2)

    # 提取id值
    ids1 = {item["id"] for item in data1}
    ids2 = {item["id"] for item in data2}

    # 检查id是否一致
    if ids1 != ids2:
        print(f"文件对 {prefix}_character.json 和 {prefix}_game_setting_prompt.json 的id值不一致，输出前缀：{prefix}")
    # else:
    #     print(f"文件对 {prefix}_character.json 和 {prefix}_game_setting_prompt.json 的id值一致")

print("检查完毕")
```

```python
import json
import os
import re

# 定义文件夹路径
folder_path = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/prompt"

# 定义文件名模式
pattern = re.compile(r"(\d+)_game_setting_prompt.json$")

# 检查文件夹是否存在
if not os.path.exists(folder_path):
    print(f"文件夹 {folder_path} 不存在，请检查路径是否正确！")
else:
    # 获取文件夹中的所有文件
    files = os.listdir(folder_path)

    # 遍历文件夹中的所有文件
    for file_name in files:
        # 使用正则表达式匹配文件名
        match = pattern.match(file_name)
        if match:
            file_path = os.path.join(folder_path, file_name)

            # 读取JSON文件
            with open(file_path, "r", encoding="utf-8") as file:
                data = json.load(file)

            # 提取trial字段的最大值
            try:
                max_trial = max(item["trial"] for item in data)
            except KeyError:
                print(f"文件 {file_name} 中缺少 'trial' 字段，无法进行检查")
                continue

            # 检查最大值是否为60
            if max_trial != 60:
                print(f"文件 {file_name} 中 item['trial'] 的最大值为 {max_trial}，不符合要求")
            # else:
            #     print(f"文件 {file_name} 中 item['trial'] 的最大值为60，符合要求")

print("检查完毕")
```

```python
import os

# 定义文件夹路径
folder_path = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/result_gpt-3.5-turbo-0125"

# 检查文件夹是否存在
if not os.path.exists(folder_path):
    print("指定的文件夹路径不存在，请检查路径是否正确！")
else:
    # 获取文件夹中的所有文件
    files = os.listdir(folder_path)
    
    # 提取文件名中的数字（假设文件名格式是 `output_数字.txt`）
    numbers = set()
    for file_name in files:
        if file_name.startswith("output_") and file_name.endswith(".txt"):
            try:
                num = int(file_name.split("_")[1].split(".")[0])
                numbers.add(num)
            except (IndexError, ValueError):
                continue  # 忽略不符合命名规则的文件
    
    # 检查0-1016中缺失的数字
    expected_numbers = set(range(0, 1017))  # 0到1016
    missing_numbers = sorted(expected_numbers - numbers)
    
    print(f"缺失的数字（0-1016）：{missing_numbers}")
    print(f"共缺失 {len(missing_numbers)} 个数字")
```

```python
import os
import json
import pandas as pd
from glob import glob

# 定义文件夹路径
prompt_folder = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/prompt"
result_folder = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/result_gpt-3.5-turbo-0125"
output_folder = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/merged"

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 获取所有prompt文件
prompt_files = glob(os.path.join(prompt_folder, "*_game_setting_prompt.json"))

# 遍历每个prompt文件
for prompt_file in prompt_files:
    # 提取prefix
    prefix = os.path.basename(prompt_file).split('_')[0]
    
    # 构造对应的result文件路径
    result_file = os.path.join(result_folder, f"output_{prefix}.txt")
    
    # 检查result文件是否存在
    if not os.path.exists(result_file):
        print(f"Result file for prefix {prefix} does not exist. Skipping...")
        continue
    
    # 读取prompt文件内容并转换为DataFrame
    with open(prompt_file, 'r', encoding='utf-8') as f:
        prompt_data = json.load(f)
    prompt_df = pd.DataFrame(prompt_data)

    # 添加prefix列
    prompt_df['prefix'] = prefix
    
    # 读取result文件内容
    with open(result_file, 'r', encoding='utf-8') as f:
        result_df = pd.read_csv(result_file, sep='\t')

    merged_df = pd.concat([prompt_df, result_df], axis=1)

    # 保存合并后的数据到新的txt文件
    output_file = os.path.join(output_folder, f"merged_{prefix}.txt")
    merged_df.to_csv(output_file, sep='\t', index=False, encoding='utf-8')
    print(f"Merged data for prefix {prefix} saved to {output_file}")
```

```python
import os
import pandas as pd
from glob import glob

folder_path = r"C:/Users/Thinkpadx13/Desktop/agent-trust-multiple/agent-trust-raw/agent_trust/merged"

# 获取该文件夹下所有txt文件，按文件名排序
txt_files = sorted(glob(os.path.join(folder_path, "*.txt")))

all_dfs = []

for file in txt_files:
    # 读取txt文件，假设是以制表符分割，且有表头
    df = pd.read_csv(file, sep='\t', header=0)
    
    # 如果文件中有多余的index列（名字一般是Unnamed: 0），就删除
    if 'index' in df.columns:
        df = df.drop(columns=['index'])
    
    all_dfs.append(df)

# 按顺序合并所有DataFrame
merged_df = pd.concat(all_dfs, ignore_index=True)

# 保存为新的csv或txt文件，默认不保存索引列
output_file = os.path.join(folder_path, "merged_all.txt")
merged_df.to_csv(output_file, sep='\t', index=False)

print(f"合并完成，保存到 {output_file}")
```
